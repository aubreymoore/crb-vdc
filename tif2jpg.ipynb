# tif2jpg

Converts a files from geoTIIF to jpg

```python
from osgeo import gdal
import os
import glob
```

```python
TIF_DIR = 'tiles'
JPG_DIR = 'jpg'
```

```python
os.makedirs(JPG_DIR, exist_ok = True)     # Make output directory if it does not already exist
kwargs_gdal = {'format': 'JPEG'}          # Set options for gdal.Translate
tif_list = glob.glob(f'{TIF_DIR}/*.tif')  # Create list of tif files 
for f in tif_list:
    input_fn = f
    output_fn = f.replace(TIF_DIR, JPG_DIR).replace('tif', 'jpg')
    gdal.Translate(output_fn, input_fn, **kwargs_gdal)
print('FINISHED')
```
